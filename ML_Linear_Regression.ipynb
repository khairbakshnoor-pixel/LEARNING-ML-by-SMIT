```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
hours = np.array([1,2,3,4,5,6])
marks = np.array([30,40,60,70,80])
```

```python
print(hours)
```

```python
X = hours.reshape(-1,1) #converts array to 2D for input to LR model
y = marks
```

```python
X
```

```python
model = LinearRegression() #create model
```

```python
#Model training
model.fit(X,y)
#Internally: Finds slope, Finds intercept, minimizes MSE
```

```python
y_pred = model.predict(X)
y_pred
```

```python
model.predict([[7]])
```

```python
model.predict([[9]])
```

```python
print('Slope',model.coef_)
print('Intercept', model.intercept_)
```

Linear regression: tries to draw a straight line that best fits the data.

Regression means finding a relationship between input and output

Linear regression tries to draw a straight line covering maximum points or maintaining least distance from every point

Equation of line y=mx+c

y: output (predicted value)
x: input
m: slope (rate of change)
c: y-intercept: the point where the line crosses y-axis: starting point

Not every point lies perfectly on the line
so we need a way to measure how wrong we are.
This is called error.

Error = (y_actual - y_predicted)

Squared Error = (y_actual - y_predicted)^2

We square the error so negative and positive errors dont cancel out.

MSE: Mean Squared Error: [SUM(squared error)]/n

The best line is the one that minimizes MSE
