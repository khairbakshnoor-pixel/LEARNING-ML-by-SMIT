Main reason of overfitting is due to the coefficient becoming too large and aggressive

Regularization: is a technique used to reduce overfitting by penalizing large coefficients.

Types of regularization

1. Ridge Regression (L2 Regularization)

MSE: Measures Prediction error
Lambda / Alpha: controls how strict the penalty is

2. Lasso Regression (L1 Regularization)

y = 3x1 + 5x2 + 90x3 + c
y = 3x1 + 5x2 + 0 + c

Lasso can actually force some coefficients to become exactly zero

That means it can completely remove features.

Lasso has Feature selection concept

the model automatically decides  which features matter, which features are useless.


regularization is highly sensitive to feature scale.

if one feature has value in thousands and another feature has values in single digits, the penalty becomes unfair.

alpha and degree etc are called hyperparameters. these are the values which are set before the training starts
