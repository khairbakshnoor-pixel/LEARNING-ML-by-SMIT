```python
# 1.Analyze survival based on gender



import matplotlib.pyplot as plt
import seaborn as sns

survival_rate_by_gender = df.groupby('Sex')['Survived'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=survival_rate_by_gender, x='Sex', y='Survived', palette='viridis')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# 2. Analyze survival based on passenger class



import matplotlib.pyplot as plt
import seaborn as sns

survival_Pcalss = df.groupby('Pclass')['Survived'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=survival_Pcalss, x='Pclass', y='Survived', hue='Pclass', palette='viridis', legend=False)
plt.title('Survival Based on Pclass')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


bins = [0, 12, 18, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Elderly']


df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


age_distribution = df.groupby('AgeGroup')['Survived'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=age_distribution, x='AgeGroup', y='Survived', hue='AgeGroup', palette='viridis', legend=False)
plt.title('Age Distribution')
plt.xlabel('AgeGroup')
plt.ylabel('Survival rate')
plt.show()
```
