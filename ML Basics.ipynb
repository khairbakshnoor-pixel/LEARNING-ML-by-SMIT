```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = np.array([1,2,3,4,5,6])
marks = np.array([30,40,50,60,70,80])

```

```python
plt.scatter(hours, marks)
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.title('Marks vsStudy Hrs')
plt.show()
```
