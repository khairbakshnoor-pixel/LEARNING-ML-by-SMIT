```python
import pandas as pd
df = pd.read_csv('heart.csv')
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
scores = cross_val_score(model, X_scaled, y, cv=5)
```

```python
print(scores)
```

```python
print(scores.mean())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression()
lr_scores = cross_val_score(lr, X_scaled, y, cv=5)
```

```python
knn = KNeighborsClassifier()
knn_scores = cross_val_score(knn, X_scaled, y, cv=5)
```

```python
dt = DecisionTreeClassifier()
dt_scores = cross_val_score(dt, X_scaled, y, cv=5)
```

```python
rf = RandomForestClassifier()
rf_scores = cross_val_score(rf, X_scaled, y, cv=5)
```

```python
svm = SVC()
svm_scores = cross_val_score(svm, X_scaled, y, cv=5)
```

```python
results = pd.DataFrame({
    'Model': [
        'Logistic Regression',
        'KNN',
        'Decision Tree',
        'Random Forest',
        'SVM'
    ],
    'CV Score': [
        lr_scores.mean(),
        knn_scores.mean(),
        dt_scores.mean(),
        rf_scores.mean(),
        svm_scores.mean()
    ]})

results
```

```python
#Grid Search

from sklearn.model_selection import GridSearchCV
```

```python
pg = {
    'n_estimators': [50,100,200],
    'max_depth': [3,5,10]
}
```

```python
grid = GridSearchCV(
    estimator = RandomForestClassifier(random_state=42),
    param_grid = pg,
    cv=5, scoring='accuracy')
```

```python
grid.fit(X, y)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

One train-test split can accidentally produce:

an easy test set
a difficult test set
a biased test set

Cross Validation solves this problem

Instead of evaluating the model once, we evaluate it multiple times.

K-Fold Cross Validation:

1025 patients
k = 5

Fold 1: 205 patient
Fold 2: 205 patient
Fold 3: 205 patient
Fold 4: 205 patient
Fold 5: 205 patient

Round 1:

Testing: Fold 1
Training: Fold 2,3,4,5

Round 2:
Testing: Fold 2
Training: Fold 1,3,4,5

Round 3... 4... 5...

Now every data point gets a chance to become testing data.
