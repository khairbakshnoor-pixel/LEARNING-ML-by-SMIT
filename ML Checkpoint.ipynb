```python
import pandas as pd

# Load the dataset
file_path = '/content/KSKeWn.csv'
df = pd.read_csv(file_path)

print("Original DataFrame head:")
print(df.head())
print("\nOriginal DataFrame info:")
df.info()

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Fill null values in numerical columns with the mean
for col in numerical_cols:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)

# Fill null values in categorical columns with the mode
for col in categorical_cols:
    if df[col].isnull().any():
        # .mode()[0] handles cases where there might be multiple modes
        df[col].fillna(df[col].mode()[0], inplace=True)

print("\nDataFrame head after filling null values:")
print(df.head())
print("\nDataFrame info after filling null values:")
df.info()

# Display the count of null values after imputation to verify
print("\nNull values remaining after imputation:")
print(df.isnull().sum())
```
